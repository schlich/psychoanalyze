```python
import os
import pandas as pd
from IPython.display import clear_output
%load_ext autoreload
%autoreload 2
```

```python
from psychoanalyze import preprocess
from psychoanalyze.preprocess import Session, Curve
```

```python
files = os.listdir('../data/0-raw')
files.remove('.ipynb_checkpoints')
```

```python
# sessions.csv
sessions = []
curves = []
points = []
session_id = 1
curve_id = 1
for f in files:
    clear_output(wait=True)
    print(f)
    session = Session(f)
    session.clean()
    session_info = {
        'Monkey': session.monkey, 
        'Date': session.date, 
        'Filename':session.filename
    }
    session_df = pd.DataFrame([session_info])
    session_df.index.name = 'Session'
    session_curves = session_df.merge(session.split_into_curves(), on=['Monkey','Date'])
    session_curves.index.name = "Curve"
    sessions.append(session_df)
    curves.append(session_curves)
all_sessions = pd.concat(sessions, ignore_index=True,)
all_curves = pd.concat(curves, ignore_index=True).drop(columns='Filename')
all_curves.index.name = 'Curve'
```

```python
# sessions.csv
sessions = []
for f in files:
    clear_output(wait=True)
    print(f)
    session = Session(f)
    session.clean()
    sessions.append(session.data)
all_sessions = pd.concat(sessions, ignore_index=True,)
```

```python
groups = all_sessions.groupby(['Monkey','Date','Ref Amp','Ref PW','Ref Freq','Ref Dur','Active Channels','Return Channels','Comp Amp','Comp PW'])['Result'].value_counts()
```

```python
groups.unstack(fill_value=0).rename(columns={'1':'Hits','0':'Misses','2':'FAs','3':'CRs'})
```
